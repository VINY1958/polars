# **##LECTURAS 56-SQL GRAFICAS EN POLARS

```python
import polars as pl
from  polars import col
```

```python
vuelos_lf = pl.scan_parquet('/content/drive/MyDrive/polars/seccion05/vuelos/vuelos.parquet')
```

```python
vuelos_lf.collect()
```

```python
aerolinea_vs_tiempo_de_aire = vuelos_lf.group_by('AIRLINE').agg(
    AVG_AIR_TIME=col('AIR_TIME').mean()
).collect()
```

```python
aerolinea_vs_tiempo_de_aire
```

```python
vuelos_AA_DL = vuelos_lf.filter(
    (col('AIRLINE') == 'AA') | (col('AIRLINE') == 'DL')
).collect()
```

```python
vuelos_AA_DL
```

```python
hexagonal_data = vuelos_lf.select(
    col('AIRLINE'),
    col('ARRIVAL_DELAY')
).collect()
```

```python
hexagonal_data
```

```python
departure_delay_day = vuelos_lf.select(
    col('DAY'),
    col('AIRLINE'),
    col('DEPARTURE_DELAY')
).collect()


```

```python
departure_delay_day
```
